## Problema Geral: A soma de dois números digitados pelo usuário

### Abstração:
- Peça dois numeros ao usuario;
- Some esses dois numeros;
- Mostre o resultado;
- Teste e melhore o código.

```python
def somar (n1, n2):
    soma = n1 + n2
    return soma

try:
    n1 = float(input("Digite o primeiro número: "))
    n2 = float(input("Digite o segundo número: "))

    resultado = somar(n1, n2)
    print(f"A soma é: {resultado}")
except ValueError:
    print("Erro: Digite apenas números válidos!")
```
