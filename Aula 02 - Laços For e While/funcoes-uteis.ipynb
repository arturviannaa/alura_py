## Funções úteis em laços

- len() é utilizada para obter o comprimento de uma lista, string ou outro tipo de coleção. Ela nos permite saber quantas iterações precisamos realizar em um laço.
- range() gera uma sequência de números, que é frequentemente utilizada para controlar a iteração em laços for. COm ela, podemos especificar um intervalo de números para iterar, podendo também definir um passo. Por exemplo, range(6) gera os numeros de 0 a 5, permitindo que o laço for execute cinco vezes.
